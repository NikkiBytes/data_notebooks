```python

"""
Helper Functions 
"""
def http_request(req):
    print("[INFO] status code response:",req.status_code)
    print("[INFO] reason:",req.reason)
    pretty_json=json.loads(req.text)
    print("[INFO] text response:", json.dumps(pretty_json, indent=4))
```

Run the request below to gather the datasets. 

```python
# all datasets Search
all_datasets='https://veupathdb.org/veupathdb/service/record-types/dataset/searches/AllDatasets/reports/standard?reportConfig={"attributes":["primary_key","organism_prefix","project_id","eupath_release","newcategory","summary","contact","wdk_weight","version","institution","build_number_introduced","pmids_download","release_policy","short_attribution","type","genecount"],"tables":["Publications","Contacts","GenomeHistory","DatasetHistory","Version","References","HyperLinks","GeneTypeCounts","TranscriptTypeCounts"],"attributeFormat":"text"}'

all_req=requests.get(all_datasets)
datasets_dict=all_req.json()
```

```python
records_list=datasets_dict['records']
```

```python
# example record
#print(json.dumps(records_list[0], indent=4))
```

```python
def setup_df(_data, key, verbose=True):
    _df = pd.DataFrame(_data, index=[0])
    #_df.columns = ["key", "value"]

    if key ==  'tables.GenomeHistory': tables_gh.append(_df)
    elif key ==  'tables.Publications': tables_pubs.append(_df)
    elif key ==  'tables.Contacts': tables_contacts.append(_df)
    elif key ==  'tables.Version': tables_version.append(_df)
    elif key ==  'tables.GeneTypeCounts': tables_gtc.append(_df)
    elif key ==  'tables.References': tables_ref.append(_df)
    elif key ==  'tables.DatasetHistory': tables_dh.append(_df)
    elif key ==  'tables.HyperLinks': tables_hl.append(_df)
    elif key ==  'tables.TranscriptTypeCounts': tables_ttc.append(_df)
    
    
    if verbose == True:
        display(_df)
    return _df



tables_gh=[]
tables_pubs=[]
tables_contacts=[]
tables_version=[]
tables_gtc=[]
tables_ref=[]
tables_dh=[]
tables_hl=[]
tables_ttc=[]

attributes=[]

for _dict in records_list:

    _id=_dict['id'][0]['value']
    #print('[INFO] dataset id: ', _id)
    verbose=False
    if verbose==True:
        print("[INFO] keys: ", _dict.keys())
        print("[INFO] displayName: ", _dict['displayName'])
        print("[INFO] recordClassName: ", _dict['recordClassName'])
        print("[INFO] id: ", _dict['id'])
        print("[INFO] tableErrors: ", _dict['tableErrors'])

    attributes_df=pd.DataFrame(_dict['attributes'], index=[0])
    attributes.append(attributes_df)

    ############################    
    #  get available tables 
    ############################
    for key in list(_dict['tables'].keys()):
        # assign variables
        _key="tables."+key
        #print("elif key ==  '%s': tables_.append(_df)"%_key )
        temp_list=_dict['tables'][key]

        
        if len(temp_list) == 1:
            _data=temp_list[0]
            if type(_data) is dict: 
                setup_df(_data, _key, verbose=False)
            else:
                print(_data)

        elif len(temp_list) > 1:
            for _data in temp_list:
                if type(_data) is dict:
                    setup_df(_data, _key, verbose=False)
                else:
                    print(_data)

        elif len(temp_list) == 0:
            #print("[INFO] key: %s | val: %s "%(key, temp_list))
            pass

  
attr_df=pd.concat(attributes)
#display(attr_df.head())

df_gh=pd.concat(tables_gh)
#display(df_gh.tail())

df_pubs=pd.concat(tables_pubs)
#display(df_pubs.tail())

df_contacts=pd.concat(tables_contacts)
#display(df_contacts.tail())

df_version=pd.concat(tables_version)
#display(df_version.tail())

df_gtc=pd.concat(tables_gtc)
#display(df_gtc.tail())

df_ref=pd.concat(tables_ref)
#display(df_ref.tail())

df_dh=pd.concat(tables_dh)
#display(df_dh.tail())

df_hl=pd.concat(tables_hl)
#display(df_hl.tail())

df_ttc=pd.concat(tables_ttc)
#display(df_ttc.tail())


```
